```python
import random

states = ['A', 'B', 'C']
goal_state = {'A': 0, 'B': 0, 'C': 0}
actions = [0, 1]
transition_model = {
    'A': ['B'],
    'B': ['A', 'C'],
    'C': ['B']
}

initial_state = {}
path_cost = 0

print("Enter initial location (A/B/C): ")
location = input().upper()
print("Enter status of current location (0/1): ")
status = int(input())

initial_state[location] = status

for state in states:
    if state != location:
        print(f"Enter status of {state}: ")
        initial_state[state] = int(input())

def is_goal(state):
    return state == goal_state

def successors(state):
    successors = []
    for location in states:
        if state[location] == 1:
            for action in actions:
                if action == 0:

                    new_state = state.copy()
                    new_state[location] = 0
                    successors.append(('CLEAN', new_state, 1))
                else:

                    for neighbor in transition_model[location]:
                        if state[neighbor] == 1:
                            new_state = state.copy()
                            new_state[location] = 0
                            new_state[neighbor] = state[neighbor]
                            successors.append((f'MOVE {location}->{neighbor}', new_state, 1))
    return successors


state = initial_state
while not is_goal(state):

    print("Current state:", state)
    print("Path cost:", path_cost)

    available_actions = [action for action, _, _ in successors(state)]

    action, new_state, step_cost = random.choice(successors(state))
    path_cost += step_cost

    state = new_state

    print("Action:", action)

print("Goal state reached")
print("Final state:", state)
print("Path cost:", path_cost)
```
